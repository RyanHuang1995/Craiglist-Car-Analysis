```python
import pandas as pd
import sqlite3
```

```python
models = ["accord","altima","camry","civic","corolla","prius"]
```

```python
csv_file_path = [f"{model}_data.csv" for model in models]
csv_file_path
```

```python
combined_csv = pd.concat([pd.read_csv(f) for f in csv_file_path])
```

```python
combined_csv.to_csv("combined_csv.csv")
combined_df = pd.read_csv("combined_csv.csv")
conn = sqlite3.connect("Craiglist.db")
combined_df.to_sql("Craiglist Car",conn,if_exists = "replace")
```

```python
pd.read_sql_query("select * from `Craiglist Car`;", conn)
```
